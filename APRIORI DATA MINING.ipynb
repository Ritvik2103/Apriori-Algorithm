### PROCESS OF APPLYING APRIORI (COMBINATIONS OF 2)

HOW THE METHOD WORKS:

say m1=[g1,g2]
   
    m2=[g1,g3]
   
    m3=[g1,g2]
   
    m4=[g2,g3]
   
    m5=[g1,g2]
   
---Finding Frequency---

g1= 4/5(number of movies in which g1 occurs)/(total number of movies) = 80%

g2= 3/5= 60%

g3= 2/5= 40%

-- say Minimum support 30%--(i.e each genre ought to contribute a 30% view rate)

g1--> approved

g2--> approved

g3--> approved

--- Finding combinations--

{g1,g2}= 3/5 = 60%

{g2,g3}= 1/5 = 20%

{g1,g3}= 1/5 = 20%


-- Threshold Confidence say 30%--

{g1,g2} --> approved

{g2,g3} --> not approved

{g1,g3} --> not approved


-- Final Combination --

{g1,g2}   #indicating these two genres go together very well

### APPLYING APRIORI TO GENRE

FINDING THE UNIQUE GENRE

```python
gen=[]
for i in df['Genre']:
    gen+=(i.split('|')) 
gen= list(set(gen))   #gives the unique genre in the dataset
```

```python
gen, len(gen) 
```

FINDING THE NUMBER OF MOVIES FOR EACH GENRE(TO FIND OUT THE PRECENTAGE AS SHOWN IN THE ALGO)

```python
data=[]
for j in gen:
    c=0
    for i in df['Genre']:
        if(j in i):
            c+=1
    #print(j,c,round(c*100/len(df), 2))  
    #prints the genre followed by the number of movies and percentage of movies with the genre
    data.append([j, round(c*100/len(df),2)])
data= pd.DataFrame(data, columns=['Genre', 'Percentage'])
        
```

```python
data
```

```python
data.sort_values(by='Percentage', ascending= False)
```

CONSIDERING MINIMUM SUPPORT 20%

```python
data=data[data['Percentage']>20]
```

```python
data
```

CREATION OF COMBINATIONS

```python
gen_lst= list(data['Genre'])
```

```python
combinations=[]
for i in range(len(gen_lst)):
    for j in range(i, len(gen_lst)):
        if(i != j):
            #print(gen_lst[i], gen_lst[j])   prints the combinations(unique)
            combinations.append([gen_lst[i], gen_lst[j]])
```

```python
combinations
```

FINDING FREQUENCY OF EACH COMBINATION IN THE MAIN DATASET(df) 

```python
data_comb=[]
for comb in combinations:
    c=0    #acts as a counter for each combination 
    for j in df['Genre']:
        if(comb[0] in j.split('|') and comb[1] in j.split('|')):   
        #to check if the elements of the combination are present in the dataset
        
            c+=1    #counter for each occurence
    #print(comb, c)  #print the combination along with the number of movies with the genre
    data_comb.append([comb, round(c*100/len(df),2)])   #to obtain the percentage of movies with the combination
data_comb= pd.DataFrame(data_comb, columns=['Genre', 'Percent'])
        
    
```

```python
data_comb
```

APPLYING THRESHOLD OF 10%

```python
data_comb=data_comb[data_comb['Percent']>10]
```

```python
data_comb
```

PRINTING FINAL LIST OF GENRES THAT GO TOGETHER

```python
for i in data_comb['Genre']:
    print(','.join(i))
```

### APPLYING APRIORI TO DIRECTORS

```python
df
```

FINDING UNIQUE DIRECTOR

```python
direct=[]
for i in df['Director']:
    direct.extend(i.split('|'))
direct=list(set(direct))
```

```python
len(direct)
```

```python
data=[]
for dire in direct:
    c=0
    for i in df['Director']:
        if(dire in i):
            c+=1
    #print(dire, c) #prints the director with the number of movies with the director
    data.append([dire, round(c*100/len(df),2)])  #percentage of movies with director
data= pd.DataFrame(data,columns=['Director', 'Percentage'])
```

```python
data['Percentage'].describe()
```

CONSIDER MINIMUM SUPPORT 1% 

```python
data=data[data['Percentage']>0.6]
```

MAKING COMBINATIONS

```python
director= list(data['Director'])
```

```python
from tqdm import tqdm #can be used to look at the progress
```

```python
dire_comb=[]
for i in tqdm(range(len(director))):
    for j in (range(i, len(director))):
        if(i !=j):
            dire_comb.append([director[i], director[j]])
    
```

FINDING PERCENTAGE OF EACH DIRECTOR COMBINATION

```python
data_comb=[]
for comb in dire_comb:
    c=0    #acts as a counter for each combination 
    for j in df['Director']:
        if(comb[0] in j.split('|') and comb[1] in j.split('|')):   
        #to check if the elements of the combination are present in the dataset
        
            c+=1    #counter for each occurence
    #print(comb, c)  #print the combination along with the number of movies with the genre
    data_comb.append([comb, round(c*100/len(df),2)])   #to obtain the percentage of movies with the combination
data_comb= pd.DataFrame(data_comb, columns=['Director', 'Percent'])
```

APPLYING THRESHOLD OF 0.8%

```python
data_comb=data_comb[data_comb['Percent']>0.8]
```

PRINTING FINAL LIST OF DIRECTORS

```python
for i in data_comb['Director']:
    print(','.join(i))
```

FINDING POSSIBLE COMBINATIONS OF 3

```python
combinations= []
for i in range(len(unique)):
    for j in range(i, len(unique)):
        for k in range(j, len(unique)):
            if (unique[i] != unique[j] and unique[i] != unique[k] and unique[j] != unique[k]):
                combinations.append([ unique[i],unique[j],unique[k]])
```
